**Q7. What is the concept of an unbalanced matrix?**

A matrix is said to be balanced if and only if each of its cells is also balanced, where a cell is said to be balanced if the value written in that cell is strictly higher than the number of cells adjacent to it in the matrix.If a given cell has a top, bottom, left, or right cell, then those cells are considered adjacent.
